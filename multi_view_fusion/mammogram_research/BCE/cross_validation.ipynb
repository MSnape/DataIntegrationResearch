```python


# Import all necessary classes
from sklearn.metrics import confusion_matrix, roc_curve, auc, accuracy_score, roc_auc_score
from typing import Literal
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, random_split
from PIL import Image
import os
import random
import numpy as np
from upsetplot import from_contents, plot 
from torch.optim import lr_scheduler
import pandas as pd
import matplotlib.pyplot as plt 
from collections import Counter 
from sklearn.model_selection import KFold, StratifiedKFold
import warnings
import torch.optim as optim
warnings.filterwarnings("ignore", category=FutureWarning)

%load_ext autoreload
%autoreload 2

# Important local code (using autoreload to pick up local changes)
from mammogram_dataset import MammogramDataset
from utils import train_model, evaluate_model, show_confusion_matrix, print_single_view_results_table, compare_performance_all_views, print_performance_metrics
from single_view_mammogram import SingleViewMammogram
from single_view_cnn import SingleViewCNN
from late_fusion_fc import LateFusionFC
from multi_view import MultiViewDataset
from resnet_feature_extractor import ResNetFeatureExtractor
from classifier_head import ClassifierHead
from noise_dataset import RandomNoiseDataset
from late_fusion_lr import LateFusionLR

CONFIG = {
    'batch_size': 16,
    'num_classes': 1, # Binary classification
    'num_views': 2,
    'learning_rate': 0.001,
    'mv_num_epochs': 20,
    'patience': 5,
    'device': torch.device("cuda:0" if torch.cuda.is_available() else "cpu"),
    'k_folds': 5, # Number of folds for cross-validation
    'seed': 42, # For reproducibility of K-fold splits
    'class_weight_type': "default", # None,default , inv_norm or sqrt_inv_norm
    'lr_decay': True,
    'augment' : False,
    'denoise': False,    
}

def print_run_title(config):
    """
    Prints the configuration dictionary as a formatted title for a run.
    """
    print("\n" + "="*50)
    print("           ✨ MODEL TRAINING CONFIGURATION ✨")
    print("="*50)
    for key, value in config.items():
        if key == 'device':
            # Special handling for torch.device object
            print(f"  {key.replace('_', ' ').title():<20}: {value.type}")
        else:
            print(f"  {key.replace('_', ' ').title():<20}: {value}")
    print("="*50 + "\n")

# Call the function to print the values of the CONFIG to show what is being run
print_run_title(CONFIG)

# Initialize the dataset with the path to the CBIS-DDSM directory
# and the path to CSV files.
print(os.getcwd())
dataset_root = "../data/CBIS-DDSM TCIA/manifest-ZkhPvrLo5216730872708713142/"
csv_path_for_train_dataset = '../data/CBIS-DDSM TCIA/calc_case_description_train_set.csv'
csv_path_for_test_dataset = '../data/CBIS-DDSM TCIA/calc_case_description_test_set.csv'

# Get the data we need from the csv and files 
print("Training dataset: ",end='')
mammogram_calc_train_dataset = MammogramDataset(dataset_root, csv_path=csv_path_for_train_dataset)
print("Test data dataset: ",end='')
mammogram_calc_test_dataset = MammogramDataset(dataset_root, csv_path=csv_path_for_test_dataset)

def run_kfold_late_fusion_cross_validation(full_mammogram_dataset: MammogramDataset, k_folds: int = 5, num_epochs: int = 20, patience: int = 10):
    """
    Performs K-Fold cross-validation for a Late Fusion model.

    Args:
        full_mammogram_dataset (MammogramDataset): The complete MammogramDataset.
        k_folds (int): Number of folds for cross-validation.
        num_epochs (int): Number of epochs for training each fold.
        patience (int): Patience for early stopping in each fold.
        use_augmentation_denoise (bool): If True, uses CreateResnet50DatasetAugmentDenoise;
                                         if False, uses CreateResnet50DatasetAugment.
    """
    print(f"Starting {k_folds}-Fold Cross-Validation for Late Fusion Model...")

    filtered_indices = []
    labels_for_skf = []
    for i in range(len(full_mammogram_dataset)):
        entity = full_mammogram_dataset[i]
        if len(entity['images']) == 4:
            label = 0 if "BENIGN" in entity['pathology'].upper() else 1 if "MALIGNANT" in entity['pathology'].upper() else -1
            if label == -1:
                print(f"Warning: Found uncategorized label: {entity['pathology']} at index {i}. Skipping this sample.")
                continue
            filtered_indices.append(i)
            labels_for_skf.append(label)
    
    if not filtered_indices:
        raise ValueError("No valid samples found in the MammogramDataset after filtering for 4 images and valid labels.")

    labels_for_skf = np.array(labels_for_skf)
    print(f"Found {len(filtered_indices)} valid samples for K-Fold stratification.")

    skf = StratifiedKFold(n_splits=k_folds, shuffle=True, random_state=CONFIG['seed'])

    all_fold_performance_metrics = []
    all_predictions_across_folds = []
    all_true_labels_across_folds = []

    global device
    device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")

    for fold, (train_filtered_idx, val_filtered_idx) in enumerate(skf.split(np.arange(len(filtered_indices)), labels_for_skf)):
        print(f"\n--- Fold {fold+1}/{k_folds} ---")

        current_train_original_indices = [filtered_indices[i] for i in train_filtered_idx]
        current_val_original_indices = [filtered_indices[i] for i in val_filtered_idx]

        fold_train_mammogram_subset = torch.utils.data.Subset(full_mammogram_dataset, current_train_original_indices)
        fold_val_mammogram_subset = torch.utils.data.Subset(full_mammogram_dataset, current_val_original_indices)

        print("Preparing CC and MLO single-view datasets...")
        cc_single_view_train = SingleViewMammogram(
            train_dataset=fold_train_mammogram_subset,
            test_dataset=fold_val_mammogram_subset,
            view_type="CC"
        )
        if CONFIG['denoise'] == True:
            cc_single_view_train.CreateResnet50DatasetDenoiseCV()
        elif CONFIG['augment'] == True:
            cc_single_view_train.CreateResnet50DatasetAugmentCV()
        else:
            cc_single_view_train.CreateResnet50DatasetCV()


        mlo_single_view_train = SingleViewMammogram(
            train_dataset=fold_train_mammogram_subset,
            test_dataset=fold_val_mammogram_subset,
            view_type="MLO"
        )
        if CONFIG['denoise'] == True:
            mlo_single_view_train.CreateResnet50DatasetDenoiseCV()
        elif CONFIG['augment'] == True:
            mlo_single_view_train.CreateResnet50DatasetAugmentCV()
        else:
            mlo_single_view_train.CreateResnet50DatasetCV()
        

        train_multi_view_dataset = MultiViewDataset([
            cc_single_view_train.torch_dataset_train_view,
            mlo_single_view_train.torch_dataset_train_view
        ])
        val_multi_view_dataset = MultiViewDataset([
            cc_single_view_train.torch_dataset_validation_view,
            mlo_single_view_train.torch_dataset_validation_view
        ])
        
        print(f"Fold {fold+1} train multi-view dataset size: {len(train_multi_view_dataset)}")
        print(f"Fold {fold+1} validation multi-view dataset size: {len(val_multi_view_dataset)}")

        BATCH_SIZE = CONFIG['batch_size']
        train_dataloader = DataLoader(train_multi_view_dataset, batch_size=BATCH_SIZE, shuffle=True )
        val_dataloader = DataLoader(val_multi_view_dataset, batch_size=BATCH_SIZE, shuffle=False )
        test_dataloader = DataLoader(val_multi_view_dataset, batch_size=BATCH_SIZE, shuffle=False )
        dataloaders = {'train': train_dataloader, 'val': val_dataloader}
        dataset_sizes = {'train': len(train_multi_view_dataset), 'val': len(val_multi_view_dataset)}

        print("Initializing LateFusionFC model...")
        model_late_fusion = LateFusionFC(num_classes=2, num_views=2,  feature_extractor_requires_grad=False).to(device)
        
        # To mitigate class imbalance we can use weights 
        # Count class occurrences directly from the dataset
        label_counts = Counter()
        for _, label in train_multi_view_dataset:  # Dataset returns (image, label)
            label_counts.update([label.item()])  # Ensure label is a scalar (not tensor)

        num_classes = len(label_counts)
        class_counts = torch.tensor([label_counts.get(i, 0) for i in range(num_classes)], dtype=torch.float)
       
        # Compute class frequencies (normalize to sum to 1) and comput inverse of freqs
        class_frequencies = class_counts / class_counts.sum()
        inv_freqs = 1.0 / class_frequencies

        print("Inverse freqs: ", inv_freqs)
        if CONFIG["class_weight_type"]=="inv_norm" :
            #  Normalize weights (optional, for stability)  
            class_weights = inv_freqs / inv_freqs.sum()
            pos_weight = class_weights[1] / class_weights[0]
        elif CONFIG["class_weight_type"] == "sqrt_inv_norm":
            inv_freqs = torch.sqrt(inv_freqs)  # or torch.log(1 + inv_freqs)
            class_weights = inv_freqs / inv_freqs.sum()   
            pos_weight = class_weights[1] / class_weights[0]
        elif CONFIG["class_weight_type"] == "default":
            pos_weight = label_counts[0] / label_counts[1]
            pos_weight = torch.tensor([pos_weight], dtype=torch.float32) 
            print("label_counts[0] / label_counts[1]" , label_counts[0] , label_counts[1] )
        elif CONFIG["class_weight_type"] is None:
            class_weights = None
            pos_weight = None
        
        print("Class weights to mitigate class imbalance, pos_weight, : ", pos_weight)

        criterion = nn.BCEWithLogitsLoss(pos_weight)
        optimizer = torch.optim.Adam(model_late_fusion.parameters(), lr=CONFIG['learning_rate'])
        if CONFIG['lr_decay'] == False:
            scheduler = None 
        else:
            scheduler = lr_scheduler.StepLR(optimizer, step_size=7, gamma=0.1)

        print(f"Training LateFusionFC model for Fold {fold+1}...")
        
        trained_model, _, fold_predictions, fold_true_labels = train_model(
            model_late_fusion, train_dataloader,val_dataloader, val_dataloader, criterion, optimizer, device, num_epochs, patience,  scheduler  )
        
        print(f"Evaluating LateFusionFC model for Fold {fold+1}...")
        fold_accuracy, fold_performance = evaluate_model(
            trained_model, val_dataloader, device=device
        )
        
        all_fold_performance_metrics.append(fold_performance)
        all_predictions_across_folds.extend(fold_predictions)
        all_true_labels_across_folds.extend(fold_true_labels)


    print("\n--- K-Fold Cross-Validation Complete for Late Fusion Model ---")

    # Initialize avg_metrics 
    avg_metrics = {
        'Accuracy': [],
        'Precision': [],
        'Recall': [],
        'Specificity': [],
        'F1-Score': [], 
        'AUC': [],
        'TP': [],
        'FP': [],
        'TN': [],
        'FN': []
    }

    # Populate lists for averaging
    for fold_metrics in all_fold_performance_metrics:
        for key in avg_metrics.keys(): # Iterate over the keys we expect to average
            value = fold_metrics.get(key, np.nan) # Get the value, default to NaN if key is missing
            if isinstance(value, (int, float, np.integer)) and not np.isnan(value):
                avg_metrics[key].append(value)
            # If it's NaN or not a number, we just don't add it to the list for averaging

    # Calculate the actual averages
    final_avg_metrics = {}
    for key, values in avg_metrics.items():
        if values: # Only calculate mean if there are valid numbers to average
            final_avg_metrics[key] = np.mean(values)
        else:
            print(f"No values for {key} defaulted to Nan")
            final_avg_metrics[key] = np.nan # If no valid values were collected, mark as NaN

    final_avg_metrics['Name'] = "Combined Late Fusion" 

    # --- Now call print_performance_metrics with the correctly populated dictionary ---
    print("\n--- Average Performance Metrics Across All Folds (Late Fusion) ---")
    avg_tp = final_avg_metrics.get('TP', 0)
    avg_fp = final_avg_metrics.get('FP', 0)
    avg_tn = final_avg_metrics.get('TN', 0)
    avg_fn = final_avg_metrics.get('FN', 0)
    print_performance_metrics(avg_tp, avg_fp, avg_tn, avg_fn)

    print(f"Average Accuracy: {final_avg_metrics.get('Accuracy', np.nan):.4f}")
    print(f"Average Precision: {final_avg_metrics.get('Precision', np.nan):.4f}")
    print(f"Average Recall: {final_avg_metrics.get('Recall', np.nan):.4f}")
    print(f"Average F1-Score: {final_avg_metrics.get('F1-Score', np.nan):.4f}")
    print(f"Average AUC: {final_avg_metrics.get('AUC', np.nan):.4f}") 
    
    final_cm = confusion_matrix(all_true_labels_across_folds, all_predictions_across_folds)
    show_confusion_matrix(final_cm, "Combined K-Fold Confusion Matrix (Late Fusion)")

    return final_avg_metrics, final_cm

count_benign = 0
for entity in mammogram_calc_train_dataset.roi_entities:
    if "BENIGN" in (entity['pathology']).upper(): 
        count_benign += 1
count_malignant = len(mammogram_calc_train_dataset) - count_benign
print(f"For the training data, there are {count_benign} BENIGN and {count_malignant} MALIGNANT pathologies.")

full_train_pool_mammogram_ds = mammogram_calc_train_dataset
global_fixed_test_mammogram_ds = mammogram_calc_test_dataset

# Run K-Fold Cross-Validation for the Late Fusion model
kfold_performance_metrics, kfold_confusion_matrix = run_kfold_late_fusion_cross_validation(
    full_mammogram_dataset=full_train_pool_mammogram_ds,
    k_folds=CONFIG['k_folds'],
    num_epochs=CONFIG['mv_num_epochs'], 
    patience=CONFIG['patience']
)

print("\nK-Fold Cross-Validation for Late Fusion completed!")
```
